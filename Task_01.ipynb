```python
import librosa
import os
import numpy as np
import soundfile as sf
import sklearn
```

```python
path="C:\\Users\\Hp\\Downloads\\Prog AI"
l=[]
fme=[]
ok=[]
labels=[]
for i in os.listdir(path):
    fname=path+'\\' +i
    print(fname)
    for gg in os.listdir(fname):
        xx=fname+"\\"+ gg
        print(xx)
        for j in os.listdir(xx):
            data, sample_rate=librosa.load(xx+"\\"+j)
            mfc=librosa.feature.mfcc(y=data,sr=sample_rate)
            mfc=mfc.flatten().tolist()
            hop_length=512
            chromagram = librosa.feature.chroma_stft(data, sr=sample_rate, hop_length=hop_length)
            chromagram = chromagram.flatten().tolist()
            features=librosa.feature.spectral_rolloff(y=data, sr=sample_rate)
            features=features.flatten().tolist
            if(len(mfc)>600 and len(mfc)<4000):
                fme.append(features)
                ok.append(chromagram)            
                l.append(mfc)
                labels.append(i)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report 
from sklearn.neighbors import KNeighborsClassifier
from imblearn.over_sampling import SMOTE
from sklearn import svm
import pickle
```

```python
!pip install imblearn
import imblearn
from imblearn.over_sampling import SMOTE
sm = SMOTE()
df0,df1=sm.fit_resample(df0,df1)
```

```python
X_train, X_test, y_train, y_test =train_test_split(df0, df1, test_size=0.2, random_state=12)
```

```python
from sklearn.neighbors import KNeighborsClassifier #from notebook file
clf=KNeighborsClassifier()
clf.fit(X_train, y_train)
pred=clf.predict(X_test)
clf.score(X_test,y_test)
```

```python
pickle.dump(clf,open('knn.pkl','wb'))
print ("Accuracy : " , accuracy_score(y_test,pred)*100)  
print("Report : \n", classification_report(y_test, pred))
print("F1 Score : ",f1_score(y_test, pred, average='macro')*100)
```

```python
from sklearn.ensemble import RandomForestClassifier #https://www.geeksforgeeks.org/random-forest-classifier-using-scikit-learn/
RFclf = RandomForestClassifier(n_estimators = 50)
RFclf.fit(X_train, y_train)
pred = RFclf.predict(X_test)
```

```python
print ("Accuracy : " , accuracy_score(y_test,pred)*100)  
print("Report : \n", classification_report(y_test, pred))
print("F1 Score : ",f1_score(y_test, pred, average='macro')*100)
pickle.dump(clf, open("rfc.pkl", 'wb'))
```

```python
from sklearn.ensemble import ExtraTreesClassifier#https://www.tutorialspoint.com/machine_learning_with_python/machine_learning_with_python_extra_trees.html
et = ExtraTreesClassifier(n_estimators=10, max_depth=None,min_samples_split=4, random_state=0)
et.fit(X_train, y_train)
pred = et.predict(X_test)
```

```python
print ("Accuracy : " , accuracy_score(y_test,pred)*100)  
print("Report : \n", classification_report(y_test, pred))
print("F1 Score : ",f1_score(y_test, pred, average='macro')*100)
pickle.dump(et, open("et.pkl", 'wb'))
```

```python
import lightgbm as lgb #https://www.kaggle.com/prashant111/lightgbm-classifier-in-python
lg = lgb.LGBMClassifier()
lg.fit(X_train, y_train)
pred=lg.predict(X_test)
```
